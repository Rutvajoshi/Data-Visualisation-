```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data
data = pd.read_csv('/Users/rutvajoshi/Documents/Masters final project/Final submission/FinalAbarth.csv')
```

```python
data.head()
```

```python
# Group the data by 'Year' and count the comments for each year
comment_counts = data.groupby('year')['comments'].count()
```

```python
# Create a bar chart to visualize the comment count over the years
plt.figure(figsize=(10, 6))
comment_counts.plot(kind='bar', color='blue')
plt.xlabel('Year')
plt.ylabel('Comment Count')
plt.title('Comment Count Over the Years for Abarth')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a line chart to visualize the comment count over the years
plt.figure(figsize=(10, 6))
comment_counts.plot(kind='line', marker='o', color='blue')
plt.xlabel('Year')
plt.ylabel('Comment Count')
plt.title('Comment Count Over the Years For Abarth')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a line chart to visualize the comment count over the years
plt.figure(figsize=(10, 6))
comment_counts.plot(kind='line', marker='o', color='blue')
plt.xlabel('Year')
plt.ylabel('Comment Count')
plt.title('Comment Count Over the Years For Abarth')
plt.grid(True)

# Customize x-axis labels to include all years from 2013 to 2023
plt.xticks(range(2013, 2024), rotation=45)

plt.show()
```

```python
# Filter the data to include only the year 2023
data_2023 = data[data['year'] == 2023]
```

```python
# Group the filtered data by 'Month' and count the comments for each month
comment_counts = data_2023.groupby('month')['comments'].count()
```

```python
# Define month abbreviations
month_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'July']
```

```python
# Create a bar chart to visualize the comment count for each month in 2023
plt.figure(figsize=(10, 6))
comment_counts.plot(kind='bar', color='pink')
plt.xlabel('Month')
plt.ylabel('Comment Count')
plt.title('Comment Count for Each Month in 2023 for Abarth')
# Set custom x-axis tick labels for abbreviated month names
month_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun','July']
plt.xticks(range(len(month_labels)), month_labels, rotation=45)

plt.show()
```
