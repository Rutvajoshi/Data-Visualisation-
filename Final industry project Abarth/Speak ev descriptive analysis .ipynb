```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data
data = pd.read_csv('/Users/rutvajoshi/Documents/Masters final project/all_comments_new.csv')
```

```python
# Drop rows with invalid month names (optional)
data = data[data['Month'].isin(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])]
```

```python
# Replace any incorrect or alternative month names with the correct ones (optional)
# For example, replace 'January' with 'Jan', 'Sept' with 'Sep', etc.
data['Month'] = data['Month'].replace({
    'January': 'Jan',
    'February': 'Feb',
    'March': 'Mar',
    'April': 'Apr',
    'May': 'May',
    'June': 'Jun',
    'July': 'Jul',
    'August': 'Aug',
    'September': 'Sep',
    'Sept': 'Sep',
    'October': 'Oct',
    'November': 'Nov',
    'December': 'Dec'
})
```

```python
# Reset the index after dropping or modifying rows
data.reset_index(drop=True, inplace=True)
```

```python
# Print the cleaned DataFrame
print(data)
```

```python
# Create a list of month names in chronological order
months_in_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
# Calculate class distribution based on the cleaned "Month" column
class_distribution = data['Month'].value_counts()
```

```python
# Reindex the class distribution in chronological order
class_distribution = class_distribution.reindex(months_in_order, fill_value=0)
```

```python
# Print the class distribution
print("Class Distribution based on 'Month':\n")
print(class_distribution)
```

```python
# Plot the line chart
plt.figure(figsize=(10, 6))
class_distribution.plot(kind='line', marker='o', color='green')
plt.title('Class Distribution based on Month', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Group by year and prompt to calculate comment counts
comment_counts = data.groupby(['Year', 'prompt'])['comment'].count().reset_index()
```

```python
# Pivot the data to have prompts as columns and years as rows
pivot_table = comment_counts.pivot(index='Year', columns='prompt', values='comment').fillna(0)
```

```python
# Reset the index to make 'Year' a column again
pivot_table.reset_index(inplace=True)
```

```python
# Filter the DataFrame to include only comments from 2012 to 2023
filtered_data = pivot_table[(pivot_table['Year'] >= 2012) & (pivot_table['Year'] <= 2023)]
```

```python
# Create a line chart for the filtered data
filtered_data.plot(kind='line', x='Year', marker='o')
plt.title('Comment Counts Over Years for Different Prompts (2012-2023)')
plt.xlabel('Year')
plt.ylabel('Comment Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend(title='Prompt')
plt.show()
```
