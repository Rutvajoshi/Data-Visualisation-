```python
import pandas as pd
from langdetect import detect
```

```python
df = pd.read_csv('/Users/rutvajoshi/Documents/Masters final project/fiatFinaldata.csv')
```

```python
# Get the number of rows and columns
num_rows = df.shape[0]
num_columns = df.shape[1]
```

```python
# Print the number of rows and columns
print("Number of rows:", num_rows)
print("Number of columns:", num_columns)
```

```python
# Define a function to check if a sentence is in English
def is_english(sentence):
    try:
        if detect(sentence) == 'en':
            return True
        else:
            return False
    except:
        return False
```

```python
# Apply the function to filter out non-English sentences
df['is_english'] = df['Tweet Text'].apply(is_english)
df = df[df['is_english'] == True]
```

```python
# Get the number of rows and columns
num_rows = df.shape[0]
num_columns = df.shape[1]
```

```python
# Print the number of rows and columns
print("Number of rows:", num_rows)
print("Number of columns:", num_columns)
```
